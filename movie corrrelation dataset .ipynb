#### Numberize all the object columns

```python
df_numerized=Newdf.copy()
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes #this is where it actually gives the random numerization

df_numerized.head()
```

```python
Newdf.head(5)
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.title('Correlation matrix for numeric features')
plt.show()
```

```python
df_numerized.corr()
```

```python
corr_mat=df_numerized.corr()
corr_pairs=corr_mat.unstack()
corr_pairs
#unstacking what is does is , it says here's our budget and this is what all the things are compared to for budget
```

```python
sorted_pairvalues=corr_pairs.sort_values()
sorted_pairvalues
```

```python
correlation_highest=sorted_pairvalues[sorted_pairvalues > 0.5]
correlation_highest
```
